## Import packages

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Read data files

```python
df_fpkms = pd.read_csv( "../../qbb2020/data/fpkms.csv", index_col="t_name" )
df_fpkms
```

## Calculate M and A values

```python
male_fpkms = df_fpkms[df_fpkms.columns[1:9]]
male_fpkms = np.log2(male_fpkms+1)
male_fpkms
```

```python
female_fpkms = df_fpkms[df_fpkms.columns[9:]]
female_fpkms = np.log2(female_fpkms+1)
female_fpkms
```

```python
stages = ["10", "11", "12", "13", "14A", "14B", "14C", "14D"]
male_fpkms.columns = stages
female_fpkms.columns = stages
M = male_fpkms - female_fpkms
A = (male_fpkms + female_fpkms)/2
```

## Plot MA

```python
fig, ax = plt.subplots()
ax.scatter(A, M, s = 1)
ax.set_xlabel("log2(geometric mean)")
ax.set_ylabel("log2(quotient)")
ax.set_title("MA plot for male and female expression data")
```
