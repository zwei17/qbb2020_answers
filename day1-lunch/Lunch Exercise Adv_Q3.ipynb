## Import packages

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Read data

```python
df_fpkms = pd.read_csv( "../../qbb2020/data/fpkms.csv", index_col="t_name" )
df_fpkms
```

```python
df_fpkms_rep = pd.read_csv( "../../qbb2020/data/replicates.csv", index_col="t_name" )
df_fpkms_rep
```

## Extract male and female data

```python
goi = "FBtr0331261"
df_sxl = df_fpkms.loc[goi,:]
male = df_sxl[1:9]
female = df_sxl[9:]

df_sxl_rep = df_fpkms_rep.loc[goi,:]
male_rep = df_sxl_rep[:4]
female_rep = df_sxl_rep[4:-1]
```

## Plot curves of two sexes

```python
stages = ["10", "11", "12", "13", "14A", "14B", "14C", "14D"]
male.index = stages
female.index = stages
male_rep.index = stages[4:]
female_rep.index = stages[4:]

fig, ax = plt.subplots(figsize = (10.0, 8.0))
ax.plot(male.index, male, label = 'Male')
ax.plot(female.index, female, label = 'Female')
ax.plot(male_rep.index, male_rep, label = 'Male_rep')
ax.plot(female_rep.index, female_rep, label = 'Female_rep')
ax.set_xlabel("Stages")
ax.set_ylabel("fpkm")
plt.legend()
plt.show()
```
